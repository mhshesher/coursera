Flipping Image

```python
img_flip=cv.flip(img3,0) # 0 means vertical flip
img_flip2=cv.flip(img3,1) # 1 means horizontal flip
```

```python
plt.subplot(131).imshow(img3)
plt.subplot(132).imshow(img_flip)
plt.subplot(133).imshow(img_flip2)
```

```python
img_rot1=cv.rotate(img3,cv.ROTATE_90_CLOCKWISE)
img_rot2=cv.rotate(img3,cv.ROTATE_90_COUNTERCLOCKWISE)
```

```python
plt.subplot(131).imshow(img3)
plt.subplot(132).imshow(img_rot1)
plt.subplot(133).imshow(img_rot2)
```

Cropping Image

```python
img2.shape
```

```python
up=100
down=300
left=100
right=300
```

```python
img5=img2[up:down,:,:]
img6=img5[:,left:right:,:]
```

```python
plt.subplot(131).imshow(img3)
plt.subplot(132).imshow(img5)
plt.subplot(133).imshow(img6)
```

Manipulating Image Pixels

```python
img7=img3.copy()
```

```python
plt.imshow(img7)
```

```python
img8=cv.rectangle(img7,pt1=(left,up),pt2=(right,down),color=(0,255,0),thickness=5)
```
