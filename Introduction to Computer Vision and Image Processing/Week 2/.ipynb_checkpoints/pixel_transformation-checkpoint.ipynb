```python
path='/home/mehedi/github/coursera/Introduction to Computer Vision and Image Processing/Week 2/Test_Image'
```

Histograms

```python
img=cv.imread(os.path.join(path,'baboon_gray.jpg'))
```

```python
plt.imshow(img)
```

```python
hist=cv.calcHist(img,[0],None,[255],[0,255])
```

```python
type(hist)
```

```python
x=np.array(range(hist.shape[0]))
hist2=hist.T
hist3=hist2.reshape(hist2.shape[1])
plt.figure(figsize=(50,10))
plt.bar(x,hist3)
```
