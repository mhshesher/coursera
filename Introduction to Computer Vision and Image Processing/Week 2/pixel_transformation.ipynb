Intensity Transformations

```python
img.shape
```

```python
img2=np.zeros((img.shape[0],img.shape[1]),dtype=int)
img2.shape
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        img2[row][col]=(2*(img[row][col]))+1
```

```python
print(np.min(img),np.max(img))
print(np.min(img2),np.max(img2))
```

```python
def hsit_calculator(img):
    
    hist=np.zeros((np.max(img)+1),dtype=int)
    for row in range(img.shape[0]):
        for col in range(img.shape[1]):
            val=img[row][col]
            hist[val]+=1

    return hist
```

```python
hist4=hsit_calculator(img2)
```

```python
print('Histogram before transformation.')
print(hist3)
print('Histogram after transformation')
print(hist4)
```

Brightness and Contrast Adjustment

```python
img4=cv.convertScaleAbs(img,alpha=2)
```

```python
img5=cv.convertScaleAbs(img,beta=200)
```

```python
plt.subplot(131).imshow(img,cmap='gray')
plt.title('Normal')
plt.subplot(132).imshow(img4,cmap='gray')
plt.title('Contrast Increased')
plt.subplot(133).imshow(img5,cmap='gray')
plt.title('Brightness Increased')
```

```python
hist6=cv.calcHist([img4],[0],None,[256],[0,256])
```

```python
hist7=cv.calcHist([img5],[0],None,[256],[0,256])
```

```python
plt.figure(figsize=(3.87*5,2.64*5))
hist3_x=np.array(range(hist3.shape[0]))
plt.subplot(311).bar(hist3_x,hist3)
plt.title('Normal')

hist6_x=np.array(range(hist6.shape[0]))
hist6=hist6.T
hist6=hist6.reshape(hist6.shape[1])
plt.subplot(312).bar(hist6_x,hist6)
plt.title('Contrast Increased')

hist7_x=np.array(range(hist7.shape[0]))
hist7=hist7.T
hist7=hist7.reshape(hist7.shape[1])
plt.subplot(313).bar(hist7_x,hist7)
plt.title('Brightness Increased')

plt.tight_layout(pad=3.0)
```
